### 1.5: Gravitational pull

$$F=G\frac{m_1m_2}{r^2}$$

```python
moon=7.342e22
earth=5.9722e24
distance=3.85e8

print("Mass of the Moon:",moon)
print ("Mass of the earth:",earth)
print ("Distance between the Earth and the Moon:",distance)

from astropy.constants import G
print (G)

F=G*((moon*earth)/distance**2)
print ("Gravitational pull:",F)
```
