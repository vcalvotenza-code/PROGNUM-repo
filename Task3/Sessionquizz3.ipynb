### 3.2: Calculate the escape velocity

```python
import math
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radius= [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

objects = {}            #make a new dictionary for each object for their mass and radius
objects['Sun'] = {'masses':1.9891e+30, 'radius':696340}
objects['Jupiter'] = {'masses':1.8986e+27, 'radius':71492}
objects['Saturn'] = {'masses':5.6846e+26, 'radius':60268}
objects['Neptune'] = {'masses':1.0243e+26, 'radius':24764}
objects['Uranus'] = {'masses':8.6810e+25, 'radius':25559}
objects['Earth'] = {'masses':5.9736e+24, 'radius':6378.1}
objects['Venus'] = {'masses':4.8685e+24, 'radius':6051.8}
objects['Mars'] = {'masses':6.4185e+23, 'radius':3396.2}
objects['Mercury'] = {'masses':3.3022e+23, 'radius':2439.7}
objects['Moon'] = {'masses':7.349e+22, 'radius':1738.1}
objects['Pluto'] = {'masses':1.25e22, 'radius':1195}

G = 6.67e-11 #state constant
escape_velocities = {}                       #create a new empty dictionary for the escape velocities that will be filled up with the loop
print("The escape velocities and masses of the objects are:")

for name in objects:                         #for every name of the objects dictionary
    M = objects[name]['masses']              #mass 
    R = objects[name]['radius']*1000         #from km to m. radius
    v=math.sqrt(2*G*M/R)                     #the escape velocity formula for each M and R
    print(name,"'s escape velocity: ",v,"m/s.",name,f"'s mass: {M:.3g}") #.3g means showing only two digits after the dot
    escape_velocities[name] = {v}            #attach each object name to its escape velocity
print()
print("The escape velocities dictionary is: ",escape_velocities)
```

# 3.10 Area and volume of a cylinder

```python
def cylinder(r,h):
    A = 2*pi*r**2+2*pi*r*h
    V = pi*r**2*h
    return A, V
r = float(input("Input radius"))
h = float(input("Input height"))

Area, Volume = cylinder(r,h)

print(f"The area is {Area} and the volume {Volume}")

ratio=Area/Volume
print("The ratio is",ratio)
```

```python
import numpy as np

x = np.array([18.49,35.48, 35.48 , 11.34 , 7.12  , 9.82  , 23.66, 44  , 66.7  , 52.4 , 55   , 25.6 , 26.19 ])

y = np.array([ 2094.75,2320.87,1224.9,872.13,432.2,751,1410,2700,5292,2550,5253,2419,2158])

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x-x_mean)**2)

b = numerator/denominator

a = y_mean-b*x_mean

print("Slope b= ",b)
print("Intercept a= ",a)

print(f"Fitted line:{a:.3f}+{b:.3f}x")
```

# 3.22 Plotting a catenary

```python
import numpy as np
import matplotlib.pyplot as plt

#integer x values
x = list(range(-5,6)) #x values ranging from -5 to 6

#function
y = []
for value in x:
    y.append(np.cosh(value))

#plot
plt.figure()
plt.plot(x, y, marker='o', label='y = cosh(x)')

plt.title("Catenary curve ", fontsize=12)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y = cosh(x)", fontsize=12)

plt.grid()
plt.legend()

plt.show()
```
