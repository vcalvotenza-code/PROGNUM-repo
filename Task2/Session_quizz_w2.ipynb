## Question 2.12: Quadrants

```python
from math import atan2, degrees
x=float(input("Select a value for x: "))
y=float(input("Select a value for y: "))
a_r=atan2(x, y) #in radians
a_d=degrees(a_r)
print(a_d)

#Conditions
if 0 < a_d < 90:
    print("This is quadrant I")

elif 90 < a_d < 180:
    print("This is quadrant II")

elif 180 < a_d < 270:
      print("This is quadrant III")
else:
    print("This is quadrant IV")
```

The difference between atan2() and atan() 9s that atan() only gives results between -pi and pi so makes the angle with positive x axis value. But atan2() gives the correct quadrandt for the angle because the signs of both inputs are known to it.

# Question 2.15: Intesection of two lists

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

#converting the lists to Python sets

n1=set(night1)
n2=set(night2)

#using intersection() to find vaues that are common

inters = set.intersection(n1, n2)
inter = list(inters)
print("The common values of night1 and night2 are: ",inter)
```
